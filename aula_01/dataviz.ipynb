```python
# !pip install matplotlib

import matplotlib.pyplot as plt #biblioteca de dataviz
```

```python
plt.plot( #plot faz gráfico de linha
    [2, 9, 9, 2, 2], #x
    [2, 2, 9, 9, 2] #y
)
```

```python
plt.figure()
plt.plot( #plot faz gráfico de linha
    [2, 9, 9, 2, 2], #x
    [2, 2, 9, 9, 2], #y
    c='green'
)

plt.scatter( #scatter faz pontos
    [2, 9, 9, 2, 2], #x
    [2, 2, 9, 9, 2], #y
    label = "pontos do cubo",
    c='pink',
    edgecolors='black'
)
plt.title("titulo vem aqui")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(loc=3, bbox_to_anchor=(1, 0.5))
plt.tight_layout()
plt.savefig("quadrado.pdf")
plt.show()
```
